```python
# Function to assign tier labels to colleges based on 'college_df'.

def assign_tier(row):
    if row['College'] in df_college['Tier 1'].values:
        return 'Tier 1'
    elif row['College'] in df_college['Tier 2'].values:
        return 'Tier 2'
    elif row['College'] in df_college['Tier 3'].values:
        return 'Tier 3'
    else:
        return 'Unknown'
```

```python
def assign_city(row):
    try:
        if row['City'] in df_city['Metrio City'].values:
            return 1
        elif row['City'] in df_city['non-metro cities'].values:
            return 0
    except KeyError as e:
        print(f"Error: Column '{e}' not found in DataFrame.")
        return None
```

```python
# Creating  a 'metro_city' , 'Tier' column in the 'df' dataframe.
df['metro_city'] = df.apply(assign_city, axis=1)
df['Tier'] = df.apply(assign_tier, axis=1)
```

```python
import pandas as pd

# Load data from CSV
data = pd.read_csv('datacomp.csv')

# Inspect the data to confirm columns
print("Columns in the dataset:", data.columns)
print("First few rows of the dataset:")
print(data.head())

# Ensure the columns are correctly named and clean up column names if necessary
data.columns = data.columns.str.strip()  # Remove any leading/trailing whitespace from column names

# Check the column names again after cleanup
print("Cleaned columns in the dataset:", data.columns)

# Verify that the required columns exist
required_columns = ['MODEL', 'R-squared Score', 'Mean Squared Error']
for col in required_columns:
    if col not in data.columns:
        raise KeyError(f"'{col}' column not found in the dataset. Available columns: {data.columns}")

# Convert scores to numeric if they are not already
data['R-squared Score'] = pd.to_numeric(data['R-squared Score'], errors='coerce')
data['Mean Squared Error'] = pd.to_numeric(data['Mean Squared Error'], errors='coerce')

# Handle potential NaN values after conversion
data = data.dropna(subset=['R-squared Score', 'Mean Squared Error'])

# Find the best model based on highest R2 and lowest MSE
best_model = min(
    data.iterrows(), 
    key=lambda x: (x[1]['Mean Squared Error'], -x[1]['R-squared Score'])
)

# Extracting values from the result
model_name = best_model[1]['MODEL']
r2_score = best_model[1]['R-squared Score']
mse_score = best_model[1]['Mean Squared Error']

print(f"Best Model: {model_name}, R-Squared: {r2_score:.4f}, MSE: {mse_score:.2f}")
```
